```python
import rasterio
import numpy as np
import os
```

```python
import pandas as pd
```

```python
import pickle
```

```python
ds_dir = r'D:\School\Strathclyde\EF_900\Data\SentinelSatData\L2'
out_dir_name = 'time_series_dump'
target_dir_name = 'target'
target_dir = os.path.join(ds_dir, target_dir_name)
c = 0 #counter
c2 = 0 #duplicate counter
ndvi_layers_number = 0 # count the ndvi layers
ndvi_list = []
```

```python
for item in os.listdir(target_dir):
    #print(item)
    if item.split('_')[2] == 'B04': #set band 4(red) as item
        c += 1
        c2 = 0
        for item2 in os.listdir(target_dir):
            #find the respective band 8
            if item2 != item and item.split("_")[1] == item2.split("_")[1] and item.split("_")[0] == item2.split("_")[0] and item2.split("_")[2] == 'B08' and (('DOUBLE' in item and 'DOUBLE' in item2) or ('DOUBLE' not in item and 'DOUBLE' not in item2)):
                c2 += 1
#                 print(f'NIR band is = {item2}')
#                 print(f'RED band is = {item}')
#                 print(f'{target_dir}\\{item}')
                red_band = rasterio.open(f'{target_dir}\\{item}')
                nir_band = rasterio.open(f'{target_dir}\\{item2}')
                red = red_band.read(1).astype('float32')
                nir = nir_band.read(1).astype('float32')
                ndvi = np.where((nir + red) == 0., np.nan, (nir - red) / (nir + red))
                ndvi_list.append(ndvi)
#                 if 'DOUBLE' in item:
#                     output_dir = f'{out_dir_name}\\{item[:19]}{item[-11:-4]}_ndvi.tif'
#                 else:
#                     output_dir = f'{out_dir_name}\\{item[:-12]}_ndvi.tif'
                
#                 output_dir = os.path.join(ds_dir, output_dir)
#                 print(f'working on ... {output_dir}')
#                 ndvi_image = rasterio.open(output_dir, 'w', driver='Gtiff',
#                                           width = red_band.width,
#                                           height = red_band.height,
#                                           count = 1, crs = red_band.crs,
#                                           transform = red_band.transform,
#                                           dtype = 'float32')
#                 ndvi_image.write(ndvi,1)
#                 ndvi_image.close()
                ndvi_layers_number += 1
                
                
print(f'no of red and nir images {c}, {c2}')
print(f'no of ndvi scenes {ndvi_layers_number}')
```

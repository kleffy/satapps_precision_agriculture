```python
username = 'kleffy'
password = 'badboy12@ADL'
#url = 'https://earthexplorer.usgs.gov/'

api = SentinelAPI(username, password)
```

```python
#get area of interest
footprint = geojson_to_wkt(read_geojson(sevilla_footprint))
```

```python
#query resultset
products = api.query(footprint,
                     date=('20190301', date(2019, 7, 30)),
                     platformname='Sentinel-1',
                     #processinglevel='Level-1C',
                     #cloudcoverpercentage=(0, 30)
                    )
print(len(products))
```

```python
#downloading the data
for i in products:
    p_data = api.get_product_odata(i)
    if 'IW_GRDH' in p_data['title']:
        #api.get_product_odata(i)
        api.download(i, ds_path_sevilla)
```

```python
#convert to dataframe
products_df = api.to_dataframe(products)
```
